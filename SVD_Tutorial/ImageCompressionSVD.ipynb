```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Reading our image and setting the size and color
plt.rcParams['figure.figsize'] = [16, 8]

#The file MUST be in the same folder
A = imread('kitty.jpg')
X = np.mean(A, -1)
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
#Computing regular SVD
U, S, VT = np.linalg.svd(X, full_matrices = False)

#Extract Diagonal Singular Values from the Sigma Matrix
S = np.diag(S)

i = 0
for k in (5, 20, 100):
    #Constructing the approximate images in the setted ranks
    X_Ap = U[:,:k] @ S[0:k,:k] @ VT[:k,:]
    plt.figure(i+1)
    i += 1
    img = plt.imshow(X_Ap)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('k = ' + str(k))
    plt.show()
```
